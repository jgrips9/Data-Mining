```python
ticker = []
accessionNo = []
description = []
filedAt = []
ciks = []
urls = []
name = []
for i in filings["filings"]:
    ticker.append(i["ticker"])
    description.append(i["description"])
    accessionNo.append(i["accessionNo"])
    name.append(i["companyNameLong"])
    filedAt.append(i["filedAt"])
    ciks.append(i["cik"])
    urls.append(i["filingUrl"])
```
